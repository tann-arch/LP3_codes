```python
# Importing libraries
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Defining the function and its derivative
def f(x):
    return (x + 5)**2

def df(x):
    return 2 * (x + 5)
```

```python
# Implementing Gradient Descent
x = 3                # Starting point
lr = 0.1             # Learning rate
precision = 0.0001   # Stop condition
max_iters = 100      # Safety stop
x_history = [x]

for i in range(max_iters):
    grad = df(x)
    new_x = x - lr * grad
    if abs(new_x - x) < precision:
        break
    x = new_x
    x_history.append(x)

print(f"Local minima occurs at x = {x:.4f}")
print(f"Number of iterations: {len(x_history)}")
```

```python
# Visualizing the Gradient Descent Path
x_vals = np.linspace(-10, 5, 100)
y_vals = f(x_vals)

plt.figure(figsize=(8,5))
plt.plot(x_vals, y_vals, label='y = (x+5)^2')
plt.scatter(x_history, [f(i) for i in x_history], color='red', label='Descent path')
plt.title('Gradient Descent to Find Local Minima')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.show()
```
